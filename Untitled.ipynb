```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('cat3.JPG')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
t = np.array([[-1,0,-1],
              [0,5,0],
              [-1,0,-1]])
result = cv2.filter2D(img, -1, t)
result = np.hstack((img, result))
plt.imshow(result)
print(img.shape)

result = cv2.cvtColor(result, cv2.COLOR_BGR2RGB)
cv2.imwrite('outl.jpg', result)
```
